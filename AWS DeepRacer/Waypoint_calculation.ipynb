```python
import numpy as np
import matplotlib.pyplot as plt

# waypoints data
data = np.load('reInvent2019_wide.npy')

# Extract columns for x, y, and labels
x = data[:, 0]
y = data[:, 1]
labels = np.arange(len(data))  # Generating numerical labels for each waypoint

# scatter plot with numerical labels
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='red')  # Plot points
for i, label in enumerate(labels):
    plt.text(x[i], y[i], str(label), fontsize=9, ha='right')

plt.title('Waypoints Visualization with Numerical Labels')
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')
plt.grid(True)
plt.show()
```
